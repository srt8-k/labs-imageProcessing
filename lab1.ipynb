```python
import cv2
import matplotlib.pyplot as plt


img = cv2.imread('image/cameraman.tif')


img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display image
plt.imshow(img_rgb)
plt.title("Cameraman - OpenCV")
plt.axis('off')
plt.show()


```

```python
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.cm as cm


img2 = Image.open('image/lena_gray_256.tif')


plt.imshow(img2, cmap=cm.Greys_r)
plt.title("Lena - PIL")
plt.axis('off')
plt.show()
```
