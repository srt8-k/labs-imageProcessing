```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('image3/cameramann.jpg') 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis('off')

```

```python
#task1.A 
```

```python
scaled_img = cv2.resize(img_rgb, None, fx=2, fy=2)

plt.imshow(scaled_img)
plt.title("Scaled Image")
plt.axis('off')

```

```python
#task1.B 
```

```python
(h, w) = img.shape[:2]
center = (w // 2, h // 2)

rotation_matrix = cv2.getRotationMatrix2D(center, 120, 1)
rotated_img = cv2.warpAffine(img_rgb, rotation_matrix, (w, h))

plt.imshow(rotated_img)
plt.title("Rotated Image (120°)")
plt.axis('off')

```

```python
#Task1.C
```

```python
shear_matrix = np.float32([
    [1, 0.5, 0],   # X shear
    [0, 1, 0]
])

sheared_img = cv2.warpAffine(img_rgb, shear_matrix, (w, h))

plt.imshow(sheared_img)
plt.title("Sheared Image")
plt.axis('off')
```

```python
#Task2.A
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
```

```python
#Task2.B
```

```python
negative = 255 - gray

plt.imshow(negative, cmap='gray')
plt.title("Negative Image")
plt.axis('off')
```

```python
#Task2.C
```

```python
gray_float = gray.astype(np.float32)

log_img = np.log1p(gray_float)

log_img = (log_img / np.max(log_img)) * 255

log_img = np.uint8(log_img)

plt.imshow(log_img, cmap='gray')
plt.title("Log Transformation")
plt.axis('off')

```

```python
#Task2.D
```

```python
gamma = 0.5   
gamma_img = np.power(gray / 255.0, gamma)
gamma_img = np.uint8(gamma_img * 255)

plt.imshow(gamma_img, cmap='gray')
plt.title("Gamma Correction")
plt.axis('off')
```

```python
cv2.imwrite('image3/negative.jpg', negative)
cv2.imwrite('image3/log.jpg', log_img)
cv2.imwrite('image3/gamma.jpg', gamma_img)
```
