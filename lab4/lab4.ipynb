```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, exposure
```

```python
#task 1
```

```python
img = data.moon()  
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title("Original Moon Image")
plt.show()
```

```python
p3, p80 = np.percentile(img, (3, 80))
print("3% percentile:", p3)
print("80% percentile:", p80)
```

```python
img_rescale = exposure.rescale_intensity(img, in_range=(p3, p80))
```

```python
fig = plt.figure(figsize=(12, 10))


fig.add_subplot(2, 2, 1)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title('Original Moon Image')

fig.add_subplot(2, 2, 2)
plt.hist(img.flat, bins=256, range=(0, 255))
plt.title("Histogram - Original")
plt.xlabel("Gray level values")
plt.ylabel("Number of pixels")

fig.add_subplot(2, 2, 3)
plt.imshow(img_rescale, cmap='gray')
plt.axis('off')
plt.title('Rescaled Moon Image (3%-80%)')

fig.add_subplot(2, 2, 4)
plt.hist(img_rescale.flat, bins=256, range=(0, 255))
plt.title("Histogram - Rescaled")
plt.xlabel("Gray level values")
plt.ylabel("Number of pixels")

plt.tight_layout()
plt.show()
```

```python
img = data.moon()
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title("Original Moon Image")
plt.show()
```

```python
img_eq = exposure.equalize_hist(img)
img_eq_255 = (img_eq * 255).astype(np.uint8)
```

```python
fig = plt.figure(figsize=(12, 8))


fig.add_subplot(1, 2, 1)
plt.imshow(img_eq, cmap='gray')
plt.axis('off')
plt.title('Histogram Equalized Image')


fig.add_subplot(1, 2, 2)
plt.hist(img_eq.flat, bins=256, range=(0,1))
plt.title('Histogram after Equalization')
plt.xlabel('Gray level values')
plt.ylabel('Number of pixels')

plt.show()
```

```python
ref_img = data.rocket()
source_img = data.chelsea()

matched_img = exposure.match_histograms(source_img, ref_img, channel_axis=-1)

fig, axes = plt.subplots(1, 3, figsize=(18, 6))
axes[0].imshow(source_img)
axes[0].axis('off')
axes[0].set_title('Original Chelsea')

axes[1].imshow(ref_img)
axes[1].axis('off')
axes[1].set_title('Reference Rocket')

axes[2].imshow(matched_img)
axes[2].axis('off')
axes[2].set_title('Chelsea after Histogram Matching')

plt.show()
```
