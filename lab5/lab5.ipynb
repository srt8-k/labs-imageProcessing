```python
# Task 1: Box Filter (Smoothing)


import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read image
image = cv2.imread(r'C:\Users\amria\image-p\lab5\Parrot.png')

if image is None:
    print("❌ Error:")
else:
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Apply 7x7 box filter
    smoothed = cv2.blur(image, (7, 7))

    # Plot
    fig, ax = plt.subplots(1, 2, figsize=(10,5))

    ax[0].imshow(image)
    ax[0].set_title("Original Image")
    ax[0].axis("off")

    ax[1].imshow(smoothed)
    ax[1].set_title("Smoothed (7x7 Box Filter)")
    ax[1].axis("off")

    plt.show()
```

```python
# Task 2: Gaussian Filters


# Read image
image = cv2.imread('images/Parrot.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Gaussian filters
gaussian_5 = cv2.GaussianBlur(image, (5, 5), 0)
gaussian_21 = cv2.GaussianBlur(image, (21, 21), 0)

# Plot
fig, ax = plt.subplots(1, 3, figsize=(15,5))

ax[0].imshow(image)
ax[0].set_title("Original Image")
ax[0].axis("off")

ax[1].imshow(gaussian_5)
ax[1].set_title("Gaussian 5x5")
ax[1].axis("off")

ax[2].imshow(gaussian_21)
ax[2].set_title("Gaussian 21x21")
ax[2].axis("off")

plt.show()
```
